```python
import numpy as np
price=np.array([20,28, 39, 51, 75, 130, 210,400])
size=np.array([100,150,200,250,500, 1000,2500,5000])
print(price,size)
```

```python
import matplotlib.pyplot as plt
plt.scatter(size,price,color="red")
linear_model=np.polyfit(size,price,1)
predicted =np.poly1d(linear_model)
print(predicted)
plt.plot(size,predicted(size))
```

```python
def cost(x,y,w):
  yhat=w*x
  J=np.dot((yhat-y).T,(yhat-y))
  return yhat, J/2*y.shape[0]
```

```python
plotJ=np.array([0,0])
for i in range(167):
  y, J=cost(size, price, i/1000)
  plotJ=np.vstack((plotJ,[i/1000,J]))
  #print(i/100,y,J)
plotJ
plt.plot(plotJ[1:,0], plotJ[1:,1])
minLoss=np.min(plotJ[1:,1])
print(minLoss, np.where(plotJ[1:,1] == minLoss))
```
